# Assessment 2: Find the closest numbers 

You have a list of numbers, and you want to find the two (nonidentical) numbers that are closest to each other. That is, the two numbers with the smallest absolute difference.

For example, let's start with a simple list of 6 numbers:

N = [ 121, 7.99, 44, 6, 2.5, 9]

When we look at this list we can deduce that the two closest numbers are 7.99 and 9. These two numbers have a difference of 1.01.

### Part 1. Plan and create the initial algorithm [40 Marks]
Develop an algorithm for determining the closest pair of numbers of a given 1D list or array. 

- Use the above list **N** as an example

- Once you have devised the logic, implement your algorithm as a Python function called 'FindClosest'

- The function should accept a single argument, **N**, a given 1D list or array

- The function should return the closest pair of numbers and their absolute difference 

If your algorithm is not working, please use a replacement function (such as sorted()) to complete the reminder of the assessment.

### Part 2. Time the algorithm [10 Marks]
Apply your function to 100 randomly generated floats from 0 - 1000.

- Measure the time required for the 'FindClosest' function to perform.

- Print the time required in milliseconds with 2 decimal places

- Do not modify the 'FindClosest' function as defined in Part 1.

### Part 3. Optimise the algorithm [10 Marks]
Now you should optimise your 'FindClosest' algorithm so that it runs as fast as possible.

- Work on a copy of your original function, name this version 'FindClosestOpt'

- When ready, measure the time required for the 'FindClosestOpt' function to complete the same 100 numbers as in Part 2

- If the original 'FindClosest' algorithm cannot be optimised further, please proceed to Part 4.

### Part 4. Analyse the algorithm efficiency [15 Marks]
As we described in Lecture 10, analyse the (optimised) algorithm for different job sizes.

- Vary the size $n$ of the input list/array 

    - $n$ randomly generated floats from 0 - 1000
    - vary $n$ from 4 to 1000 floats
        
- Measure the completion time $T(n)$ 

- Visualise $T(n)$ as a graph
