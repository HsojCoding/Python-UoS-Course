# Assessment: finding prime numbers

A natural number such as 1, 2, 3, 4, 5, 6, etc..  is called a prime number (or a prime) if it can be divided evenly by just two numbers, 1 and the number itself. Alternatively, if there are more then these two divisers then it is a composite numbers.

For example, among the numbers 1 through 6:
- 1 is not a prime number because it has only one divisor, itself
- 2 is only divisible by 1 and itself, so it is prime
- 3 is only divisible by 1 and itself, so it is prime
- 4 has three divisors: 1, 2, and 4, so it is a composite
- 5 is only divisible by 1 and itself, so it is prime
- 6 has four divisors: 1, 2, 3, and 6, so it is a composite

So the numbers 2, 3, and 5 are the prime numbers, as there are no other numbers that divide them evenly. Conversely 4 and 6 are both composite numbers. Infact, every composite number $n$ has a diviser $\le \sqrt{n}$.
